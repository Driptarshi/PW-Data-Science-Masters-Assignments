## Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. *Line Plot*:
   - *Use*: Displays data points connected by straight lines, useful for visualizing trends over time.
   - *Example*: Plotting time series data such as stock prices, temperature changes, or signal variations.

2. *Scatter Plot*:
   - *Use*: Shows the relationship between two continuous variables using dots, useful for identifying correlations or patterns.
   - *Example*: Comparing heights and weights of individuals.

3. *Box Plot*:
   - *Use*: Displays the distribution of a dataset based on a five-number summary (minimum, first quartile, median, third quartile, and maximum), useful for identifying outliers and comparing distributions.
   - *Example*: Comparing the distribution of test scores across different classes.

4. *Histogram*:
   - *Use*: Shows the frequency distribution of a single variable, useful for understanding the distribution shape, central tendency, and variability.
   - *Example*: Analyzing the distribution of ages in a population.

5. *Heatmap*:
   - *Use*: Displays data in matrix form using color intensity to represent values, useful for visualizing correlations or identifying patterns in data.
   - *Example*: Visualizing the correlation matrix of multiple variables.

```python
## Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.


import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time by Event and Region")
plt.show()
```

```python
# Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.


# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot the box plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
sns.boxplot(data=titanic, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Age Distribution by Passenger Class')

sns.boxplot(data=titanic, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Fare Distribution by Passenger Class')

plt.tight_layout()
plt.show()
```

```python
# Q4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram
sns.histplot(data=diamonds, x='price', hue='cut', element='step', stat='density', common_norm=False)
plt.title('Price Distribution by Cut Quality')
plt.show()
```

```python
# Q5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot the pair plot
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Q 6. Use the "flights" dataset from seaborn to plot a heatmap.

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Number of Passengers per Month and Year')
plt.show()
```
