## Q1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

 *Matplotlib* is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used in data science, engineering, and scientific research to visualize data and make it easier to understand and interpret.

*Why is it used?*
- To create a wide variety of plots and charts.
- To visualize data for better understanding and analysis.
- To communicate data insights effectively.
- For data exploration and presentation.

*Five plots that can be plotted using the Pyplot module of Matplotlib:*
1. Line plot
2. Scatter plot
3. Bar plot
4. Histogram
5. Box plot


```python
# Q2: What is a scatter plot? Use the given code to generate data for x and y and plot a scatter plot.

# Scatter plot: A scatter plot is a type of plot that displays values for typically two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.


import numpy as np
import matplotlib.pyplot as plt

# Generating data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Creating scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
## Q3. Why is the subplot() function used? Draw four line plots using the subplot() function with the given data.

# subplot() function: The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid and control their positions.

import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Line 4')

for ax in axs.flat:
    ax.set(xlabel='X-axis', ylabel='Y-axis')

plt.tight_layout()
plt.show()
```

```python
# Q4. What is a bar plot? Why is it used? Using the given data, plot a bar plot and a horizontal bar plot.

# Bar plot: A bar plot or bar chart is a chart that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. It is used to compare different categories or groups of data.

# Why is it used?*
# To compare different categories of data.
# To show the distribution of categorical data.
# To make it easy to see trends and patterns in data.


import numpy as np
import matplotlib.pyplot as plt

# Data for bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating bar plot
plt.figure(figsize=(10, 5))
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit in USD')
plt.show()

# Creating horizontal bar plot
plt.figure(figsize=(10, 5))
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit in USD')
plt.ylabel('Company')
plt.show()
```

```python
# Q5. What is a box plot? Why is it used? Using the given data, plot a box plot.

# Box plot: A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 

# It is used to:
# Visualize the distribution of a dataset.
# Identify outliers in the data.
# Compare distributions between different datasets.

import numpy as np
import matplotlib.pyplot as plt

# Data for box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
