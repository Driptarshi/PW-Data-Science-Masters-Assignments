## Q1. What is Estimation Statistics? Explain point estimate and interval estimate.

*Estimation Statistics* is a branch of statistics that deals with estimating population parameters based on sample data. It aims to provide an estimate of an unknown parameter using a single value (point estimate) or a range of values (interval estimate).

- *Point Estimate*: A single value used to estimate a population parameter. For example, the sample mean (x̄) is a point estimate of the population mean (μ).
- *Interval Estimate*: A range of values used to estimate a population parameter. It is often expressed as a confidence interval, which provides an interval within which the parameter is expected to lie with a certain level of confidence.

```python
## Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

import numpy as np
from scipy import stats

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level=0.95):
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_score * (sample_std_dev / np.sqrt(sample_size))
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
    return confidence_interval

# Example usage
sample_mean = 50
sample_std_dev = 10
sample_size = 30
confidence_level = 0.95

print(estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level))
```

## Q3. What is Hypothesis Testing? Why is it used? State the importance of Hypothesis Testing.

*Hypothesis Testing* is a statistical method used to make decisions about population parameters based on sample data. It involves making an initial assumption (the null hypothesis), collecting data, and then determining whether the data provides sufficient evidence to reject the null hypothesis.

- *Purpose*: It is used to test the validity of a claim or hypothesis about a population parameter.
- *Importance*: Hypothesis testing is crucial in research and decision-making as it helps in determining the statistical significance of the results, thus guiding researchers and practitioners in making informed decisions.

## Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.

*Null Hypothesis (H₀)*: The average weight of male college students is equal to or less than the average weight of female college students.

*Alternative Hypothesis (H₁)*: The average weight of male college students is greater than the average weight of female college students.

```python
## Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)
    result = 'Reject Null Hypothesis' if p_value < alpha else 'Fail to Reject Null Hypothesis'
    return t_stat, p_value, result

# Example usage
sample1 = np.random.normal(70, 10, 30)  # Sample from population 1
sample2 = np.random.normal(65, 12, 30)  # Sample from population 2
alpha = 0.05

t_stat, p_value, result = two_sample_t_test(sample1, sample2, alpha)
print(f"T-statistic: {t_stat}, P-value: {p_value}, Result: {result}")
```

## Q6: What is a Null and Alternative Hypothesis? Give some examples.

- *Null Hypothesis (H₀)*: A statement asserting that there is no effect or no difference. It is the hypothesis that researchers aim to test against.
  - Example: H₀: The average weight of male college students is equal to the average weight of female college students.
- *Alternative Hypothesis (H₁)*: A statement asserting that there is an effect or a difference. It is what researchers want to prove.
  - Example: H₁: The average weight of male college students is greater than the average weight of female college students.

## Q7: Write down the steps involved in Hypothesis Testing.

1. *State the Hypotheses*: Formulate the null (H₀) and alternative (H₁) hypotheses.
2. *Set the Significance Level (α)*: Choose a significance level (commonly 0.05 or 0.01).
3. *Collect Data*: Gather sample data.
4. *Calculate the Test Statistic*: Use an appropriate statistical test to calculate the test statistic.
5. *Determine the P-value or Critical Value*: Compare the test statistic to a critical value or determine the p-value.
6. *Make a Decision*: Reject H₀ if the test statistic exceeds the critical value or if the p-value is less than α; otherwise, fail to reject H₀.
7. *Draw a Conclusion*: Conclude based on the decision made.

## Q8. Define p-value and explain its significance in Hypothesis Testing.

The *p-value* is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true. It quantifies the strength of the evidence against the null hypothesis.

- *Significance*: A low p-value (typically < 0.05) indicates strong evidence against the null hypothesis, leading to its rejection. A high p-value suggests weak evidence against the null hypothesis, resulting in a failure to reject it.

```python
## Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.


import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    result = 'Reject Null Hypothesis' if p_value < alpha else 'Fail to Reject Null Hypothesis'
    return t_stat, p_value, result

# Example usage
np.random.seed(0)
sample1 = np.random.normal(100, 15, 30)
sample2 = np.random.normal(110, 20, 30)
alpha = 0.05

t_stat, p_value, result = two_sample_t_test(sample1, sample2, alpha)
print(f"T-statistic: {t_stat}, P-value: {p_value}, Result: {result}")
```

## Q11. What is Student’s t-distribution? When to use the t-Distribution.

*Student’s t-distribution* is a probability distribution used in statistics when the sample size is small, and the population standard deviation is unknown. It is similar to the normal distribution but has heavier tails.

- *When to use*: It is used in hypothesis testing and confidence interval estimation for small sample sizes (typically n < 30) or when the population standard deviation is unknown.

## Q12. What is t-statistic? State the formula for t-statistic.

The *t-statistic* is a value used in hypothesis testing to determine if there is a significant difference between the means of two groups. It measures the size of the difference relative to the variation in the sample data.

- *Formula*: 
  \[
  t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
  \]
  where \(\bar{x}_1\) and \(\bar{x}_2\) are the sample means, \(s_1\) and \(s_2\) are the sample standard deviations, and \(n_1\) and \(n_2\) are the sample sizes.

```python
## Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.

import numpy as np
from scipy import stats

sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

# Calculate the confidence interval
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * (sample_std_dev / np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"95% Confidence Interval: {confidence_interval}")
```

```python
## Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

# Hypotheses:
# - Null Hypothesis (H₀): The mean decrease in blood pressure is 10 mmHg (μ = 10).
# - Alternative Hypothesis (H₁): The mean decrease in blood pressure is not 10 mmHg (μ ≠ 10).

#Using a one-sample t-test:

import numpy as np
from scipy import stats

sample_mean = 8
sample_std_dev = 3
sample_size = 100
population_mean = 10
alpha = 0.05

# Calculate the t-statistic
t_stat = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=sample_size-1))

result = 'Reject Null Hypothesis' if p_value < alpha else 'Fail to Reject Null Hypothesis'

print(f"T-statistic: {t_stat}, P-value: {p_value}, Result: {result}")
```

```python
## Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.

# Hypotheses:
# - Null Hypothesis (H₀): The true mean weight is 5 pounds (μ = 5).
# - Alternative Hypothesis (H₁): The true mean weight is less than 5 pounds (μ < 5).

# Using a one-sample t-test:

import numpy as np
from scipy import stats

sample_mean = 4.8
sample_std_dev = 0.5
sample_size = 25
population_mean = 5
alpha = 0.01

# Calculate the t-statistic
t_stat = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Calculate the p-value
p_value = stats.t.cdf(t_stat, df=sample_size-1)

result = 'Reject Null Hypothesis' if p_value < alpha else 'Fail to Reject Null Hypothesis'

print(f"T-statistic: {t_stat}, P-value: {p_value}, Result: {result}")
```
