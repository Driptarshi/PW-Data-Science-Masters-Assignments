## Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

*Probability Mass Function (PMF):*
The PMF is a function that gives the probability that a discrete random variable is exactly equal to some value. It applies to discrete random variables, which take on a countable number of distinct values.

Example: Consider a fair six-sided die. The PMF for rolling the die is:
\[ P(X = x) = \frac{1}{6} \quad \text{for } x = 1, 2, 3, 4, 5, 6 \]

*Probability Density Function (PDF):*
The PDF is a function that describes the likelihood of a continuous random variable to take on a particular value. The probability of the variable falling within a particular range is given by the integral of this function over that range.

Example: For a standard normal distribution, the PDF is:
\[ f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \]

## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

*Cumulative Density Function (CDF):*
The CDF of a random variable \( X \) is the function \( F(x) \) that gives the probability that \( X \) will take a value less than or equal to \( x \). It is defined for both discrete and continuous random variables.

Example: For a fair six-sided die, the CDF is:
\[ F(x) = P(X \leq x) = \begin{cases} 
0 & \text{for } x < 1 \\
\frac{1}{6} & \text{for } 1 \leq x < 2 \\
\frac{2}{6} & \text{for } 2 \leq x < 3 \\
\frac{3}{6} & \text{for } 3 \leq x < 4 \\
\frac{4}{6} & \text{for } 4 \leq x < 5 \\
\frac{5}{6} & \text{for } 5 \leq x < 6 \\
1 & \text{for } x \geq 6 
\end{cases} \]

*Usage of CDF:*
The CDF is useful because it provides a comprehensive way to describe the distribution of a random variable. It is particularly helpful in calculating probabilities for intervals and for continuous random variables.

## Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.

*Examples:*
1. Heights of adults in a population.
2. Scores on standardized tests (e.g., SAT, IQ tests).
3. Measurement errors in scientific experiments.

*Parameters:*
- *Mean (\(\mu\))*: Determines the center of the distribution. Changing \(\mu\) shifts the distribution left or right.
- *Standard Deviation (\(\sigma\))*: Determines the spread of the distribution. A larger \(\sigma\) results in a wider and flatter curve, while a smaller \(\sigma\) results in a narrower and taller curve.

## Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

*Bernoulli Distribution:*
A Bernoulli distribution is a discrete probability distribution for a random variable which has only two possible outcomes, typically labeled 1 (success) and 0 (failure), with probability \( p \) and \( 1-p \) respectively.

Example: Flipping a coin where heads = 1 (success) and tails = 0 (failure), with \( p = 0.5 \).

*Difference:*
- *Bernoulli Distribution*: Represents a single trial.
- *Binomial Distribution*: Represents the number of successes in \( n \) independent Bernoulli trials.

## Q6: Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.

To find this probability, we first calculate the z-score for 60:
\[ z = \frac{X - \mu}{\sigma} = \frac{60 - 50}{10} = 1 \]

## Q7: Explain uniform Distribution with an example.

*Uniform Distribution:*
A uniform distribution is a type of probability distribution in which all outcomes are equally likely. Every interval of the same length within the distribution's range has an equal probability.

Example: Rolling a fair six-sided die. Each outcome (1, 2, 3, 4, 5, 6) has an equal probability of \( \frac{1}{6} \).

## Q8: What is the z-score? State the importance of the z-score.

*Z-score:*
A z-score indicates how many standard deviations an element is from the mean of the distribution. It is calculated as:
\[ z = \frac{X - \mu}{\sigma} \]

*Importance:*
- Allows comparison of scores from different distributions.
- Helps identify outliers.
- Facilitates the calculation of probabilities in a standard normal distribution.

## Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

*Central Limit Theorem (CLT):*
The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size becomes large, regardless of the population's distribution.

*Significance:*
- Allows the use of normal distribution approximations for inferential statistics.
- Justifies many statistical procedures and tests.
- Helps in making predictions and decisions based on sample data.

## Q10: State the assumptions of the Central Limit Theorem.

*Assumptions:*
1. The samples must be independent.
2. The sample size should be sufficiently large (typically \( n \geq 30 \) is considered adequate).
3. The population should have a finite variance.
