```python
# Q1.
import scipy.stats as stats
import numpy as np

# Given data
mean = 50
std_dev = 5
sample_size = 30
confidence_level = 0.95

# Z-value for 95% confidence
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error
standard_error = std_dev / np.sqrt(sample_size)

# Confidence interval
margin_of_error = z_value * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval
```

```python
## Q2.

from scipy.stats import chisquare

# Observed frequencies (example values)
observed = [20, 20, 20, 20, 10, 10]  # Modify according to actual observed data

# Expected frequencies based on the given distribution
total = sum(observed)
expected = [total * 0.2, total * 0.2, total * 0.2, total * 0.2, total * 0.1, total * 0.1]

# Chi-square goodness of fit test
chi2, p = chisquare(observed, expected)
chi2, p
```

```python
# Q3.

import pandas as pd
from scipy.stats import chi2_contingency

# Data
data = [[20, 15], [10, 25], [15, 20]]
df = pd.DataFrame(data, columns=["Group A", "Group B"], index=["Outcome 1", "Outcome 2", "Outcome 3"])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(df)
chi2, p
```

```python
# Q5.

# Given data
mean = 75
std_dev = 12
sample_size = 30  # Assume sample size
confidence_level = 0.90

# Z-value for 90% confidence
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Standard error
standard_error = std_dev / np.sqrt(sample_size)

# Confidence interval
margin_of_error = z_value * standard_error
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval
```

```python
# Q6.

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Chi-square distribution
x = np.linspace(0, 30, 1000)
y = chi2.pdf(x, df)

# Plotting
plt.plot(x, y, label=f'Chi-square (df={df})')
plt.fill_between(x, y, where=(x <= 15), color='skyblue', alpha=0.5)
plt.xlabel('Chi-square value')
plt.ylabel('Probability density')
plt.title('Chi-square Distribution')
plt.legend()
plt.show()
```
